```python
import pandas as pd
```

```python
df=pd.read_csv('Cars93.csv',keep_default_na=False,na_values=['','NA'])
```

```python
x=df.drop(columns=['id','Weight'])
```

```python
y=df[['Weight']]
```

```python
cat=list(x.columns[x.dtypes=='object'])
```

```python
con=list(x.columns[x.dtypes!='object'])
```

```python
cat
```

```python
con
```

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder,StandardScaler
from sklearn.compose import ColumnTransformer
```

```python
num_pipeline=Pipeline(steps=[('impute',SimpleImputer(strategy='mean')),
                             ('scaler',StandardScaler())])
```

```python
cat_pipeline=Pipeline(steps=[('impute',SimpleImputer(strategy='most_frequent')),
                             ('ohe',OneHotEncoder(handle_unknown='ignore',sparse_output=False))])
```

```python
pre=ColumnTransformer([('num',num_pipeline,con),
                       ('cat',cat_pipeline,cat)]).set_output(transform='pandas')
```

```python
pre
```

```python
x_pre=pre.fit_transform(x)
```

```python
x_pre
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x_pre,y,test_size=0.2,random_state=60)
```

```python
xtrain
```

```python
xtest
```

```python
ytrain
```

```python
ytest
```

```python
from sklearn.linear_model import LinearRegression
model1=LinearRegression()
```

```python
model1.fit(xtrain,ytrain)
```

```python
model1.score(xtrain,ytrain)
```

```python
model1.score(xtest,ytest)
```

## Ridge model-penalty is applied on square of the coefficients

```python
from sklearn.linear_model import Ridge
model_ridge=Ridge(alpha=1)
model_ridge.fit(xtrain,ytrain)
```

```python
model_ridge.score(xtrain,ytrain)
```

```python
model_ridge.score(xtest,ytest)
```

```python
from sklearn.linear_model import Lasso
```

```python
model_lasso=Lasso(alpha=1)
model_lasso.fit(xtrain,ytrain)
```

```python
model_lasso.score(xtrain,ytrain)
```

```python
model_lasso.score(xtest,ytest)
```

## hyperparameter alphs tuning

```python
import numpy as np
```

```python
params={'alpha':np.arange(start=0.1,stop=100,step=0.1)}
```

```python
params
```

### apply grid search cv for ridge

```python
from sklearn.model_selection import GridSearchCV
rr=Ridge()
gscv1=GridSearchCV(estimator=rr,param_grid=params,cv=5,scoring='neg_mean_squared_error')
```

```python
gscv1.fit(xtrain,ytrain)
```

```python
gscv1.best_params_
```

```python
gscv1.best_score_
```

```python
best_ridge=gscv1.best_estimator_
```

```python
best_ridge
```

```python
best_ridge.score(xtrain,ytrain)
```

```python
from sklearn.model_selection import cross_val_score
scores=cross_val_score(best_ridge,xtrain,ytrain,cv=5,scoring='r2')
```
